## **Dividing Reviews into Units using Sentance Tokenization**
Sentence tokenization is the process of splitting text into individual sentences. The sentence tokenizer will not split an individual word, so the offending text, in replacement form, is preserved intact during the tokenization process. This tokenizer divides a text into a list of sentences, by using an unsupervised algorithm to build a model for abbreviation words, collocations, and words that start sentences. It must be trained on a large collection of plaintext in the target language before it can be used.



```python
from nltk import tokenize
p = df['Review Text']
units  = []
for i in range(0,7466):
  a_list = nltk.tokenize.sent_tokenize(p[i])
  units.append(a_list)

units
```

```python
topic_dict = {"Camera":camera, "Price":price, "Screen":screen, "Battery":battery, "Performance":performance, "Software":software, "Service":service}
```

## **Functions to pre-process and Map the Topics to reviews**

```python
def clean(sentence):
  sentence = sentence.lower()
  return sentence

def words(lst):
    return lst.split()

def fetchTopic(x):
    for key, values in topic_dict.items():
        for value in values:
          if x==value:
                return key
    else:
        return 0
```

```python
final = []
for i in range(0,7466):
  rev = []
  for j in range(1,95):
    if units_df.at[i,j] != None:
      units_df.at[i,j] = clean(units_df.at[i,j])
      word_lst = words(units_df.at[i,j])
      unit = []
      for k in range(0,len(word_lst)):
        unit.append(fetchTopic(word_lst[k]))
      rev.append(unit)
  final.append(rev)
```

## **Sentiment Analysis for each unit**

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
def sentiment_scores(sentence):
  
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
  
    # polarity_scores method of SentimentIntensityAnalyzer
    # oject gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
    lst = list(sentiment_dict.items())
    return lst
```

```python
sent_final = []
for i in range(0,7466):
  sent_rev = []
  for j in range(1,95):
    if units_df.at[i,j] != None:
      sent_rev.append(sentiment_scores(units_df.at[i,j]))
  sent_final.append(sent_rev)
```

## **Finding Average of Unit Sentiments to compile them for reviews**

```python
def average(i,key,dict_):
  neg = list()
  pos = list()
  neu = list()
  #Temporary list to append sentiments 
  for d in range(0,len(dict_.get(key))):
    neg_ = sent_final[i][dict_.get(key)[d]][0][1]
    pos_ = sent_final[i][dict_.get(key)[d]][2][1]
    neu_ = sent_final[i][dict_.get(key)[d]][1][1]
    neg.append(neg_)
    neu.append(neu_)
    pos.append(pos_)
  #Calculating Average
  negative = sum(neg)/len(dict_.get(key))
  positive = sum(pos)/len(dict_.get(key))
  neutral = sum(neu)/len(dict_.get(key))
  #Forming Dictionary to extract Keys 
  sent_dict = {'Negative':negative,'Positive':positive,'Neutral':neutral} 
  return max(sent_dict, key=sent_dict.get)
```

```python
#Final Dataframe creation with Dummy Values 
sent_df = pd.DataFrame()
sent_df['Review Text'] = df['Review Text']
sent_df['SENT: POSITIVITY (Camera)']= 0
sent_df['SENT: NEGATIVITY (Camera)']=0
sent_df['SENT: NEUTRAL (Camera)']=0
sent_df['SENT: POSITIVITY (Price)']=0
sent_df['SENT: NEGATIVITY (Price)']=0
sent_df['SENT: NEUTRAL (Price)']=0
sent_df['SENT: POSITIVITY (Screen and Display)']=0
sent_df['SENT: NEGATIVITY (Screen and Display)']=0
sent_df['SENT: NEUTRAL (Screen and Display)']=0
sent_df['SENT: POSITIVITY (Battery)']=0
sent_df['SENT: NEGATIVITY (Battery)']=0
sent_df['SENT: NEUTRAL (Battery)']=0
sent_df['SENT: POSITIVITY (Processor)']=0
sent_df['SENT: NEGATIVITY (Processor)']=0
sent_df['SENT: NEUTRAL (Processor)']=0
sent_df['SENT: POSITIVITY (Software)']=0
sent_df['SENT: NEGATIVITY (Software)']=0
sent_df['SENT: NEUTRAL (Software)']=0
sent_df['SENT: POSITIVITY (Service)']=0
sent_df['SENT: NEGATIVITY (Service)']=0
sent_df['SENT: NEUTRAL (Service)']=0

#Looping through all reviews 
for i in range(0,7466):
  top_dict = {"Camera":[], 0:[], "Price":[], "Screen":[], "Battery":[], "Performance":[], "Software":[], "Service":[]}
  for a in range(0,len(final[i])): #Looping through one review, basically a = unit
    for k in range(0,len(final[i][a])): #Looping through each value of a unit, basically k = 0 OR Topics 
      top_dict[final[i][a][k]].append(a) #Position of unit appended in dictionary 
    lst = list(top_dict.values())
    #To remove repitition of values of Keys from top_dict
    flag = []
    for r in range(0,len(lst)):
      flag1 = list(set(lst[r]))
      flag.append(flag1)
      #Updated the dictionary with unique values. 
    top_dict = {"Camera": flag[0], 0: flag[1], "Price": flag[2], "Screen": flag[3], "Battery": flag[4], "Performance": flag[5], "Software": flag[6], "Service": flag[7]}

    #Final Updation of Dataframe with values 1 and 0 
    for key,values in topic_dict.items():
      if top_dict.get(key) != []:
        if key == 'Price':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Price)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Price)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Price)'] = 1
        elif key == 'Battery':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Battery)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Battery)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Battery)'] = 1
        elif key == 'Camera':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Camera)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Camera)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Camera)'] = 1
        elif key == 'Screen':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Screen and Display)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Screen and Display)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Screen and Display)'] = 1
        elif key == 'Performance':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Processor)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Processor)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Processor)'] = 1
        elif key == 'Software':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Software)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Software)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Software)'] = 1
        elif key == 'Service':
          sentiment = average(i,key, top_dict)
          if sentiment == 'Neutral':
            sent_df.at[i,'SENT: NEUTRAL (Service)'] = 1
          elif sentiment == 'Positive':
            sent_df.at[i,'SENT: POSITIVITY (Service)'] = 1
          elif sentiment == 'Negative':
            sent_df.at[i,'SENT: NEGATIVITY (Service)'] = 1

```
